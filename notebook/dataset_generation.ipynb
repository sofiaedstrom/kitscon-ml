```python
import matplotlib.pyplot as plt
import pandas as pd
import sklearn as sklearn
from sklearn import datasets
```

## Generate datasets

### Generate noisy circles with sklearn

```python
circle_data, labels = datasets.make_circles(n_samples=1000, noise=0.16, factor=0.3)
transposed_matrix = circle_data.transpose()
labeled_circle_data = {'x': transposed_matrix[0], 'y': transposed_matrix[1], 'label': labels}
circle_data_noise_df = pd.DataFrame(labeled_circle_data)
```

```python
class_zero = circle_data_noise_df[circle_data_noise_df['label'] == 0]
class_one = circle_data_noise_df[circle_data_noise_df['label'] == 1]
ax = class_zero.plot.scatter('x','y', c='crimson')
class_one.plot.scatter('x', 'y', c='teal', ax=ax)
```

```python
with open('datasets/nonlinear_circle_data_with_noise.csv', 'w') as f:
    circle_data_noise_df.to_csv(f, index=False)
```

### Generate circles with less noise

```python
circle_data, labels = datasets.make_circles(n_samples=1000, noise=0.05, factor=0.3)
transposed_matrix = circle_data.transpose()
labeled_circle_data = {'x': transposed_matrix[0], 'y': transposed_matrix[1], 'label': labels}
circle_data_df = pd.DataFrame(labeled_circle_data)
```

```python
class_zero = circle_data_df[circle_data_df['label'] == 0]
class_one = circle_data_df[circle_data_df['label'] == 1]
ax = class_zero.plot.scatter('x','y', c='crimson')
class_one.plot.scatter('x', 'y', c='teal', ax=ax)
```

```python
with open('datasets/nonlinear_circle_data.csv', 'w') as f:
    circle_data_df.to_csv(f, index=False)
```

### Generate 2D clusters

```python
cluster_data, labels = datasets.make_blobs(centers=2, n_samples=1000, n_features=2, cluster_std=1.2)
transposed_matrix = cluster_data.transpose()
labeled_cluster_data = {'x': transposed_matrix[0], 'y': transposed_matrix[1], 'label': labels}
cluster_data_df = pd.DataFrame(labeled_cluster_data)
```

```python
cluster_data_df.plot.scatter('x', 'y')
```

```python
with open('datasets/2d_cluster_data.csv', 'w') as f:
    cluster_data_df.to_csv(f, index=False)
```
